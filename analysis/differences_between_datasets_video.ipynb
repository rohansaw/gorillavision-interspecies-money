```python
# Visualize them with umap and pca
# %pip install pandas matplotlib datashader bokeh holoviews scikit-image colorcet
# %pip install  umap-learn[plot]
# % 
import umap
import numpy as np
import umap.plot
import numpy as np
```

```python
bwindi_embeddings = np.load('/workspace/data/crops_and_embeddings/martha_bwindi/embeddings/embeddings.npy')
loango_embeddings = np.load('/workspace/data/crops_and_embeddings/martha_loango/embeddings/embeddings.npy')
kwitonda_embeddings = np.load('/workspace/data/crops_and_embeddings/kwitonda_portraits/embeddings/embeddings.npy')
kwitonda_database_embeddings = np.load('/workspace/data/crops_and_embeddings/kwitonda_database/embeddings/embeddings.npy')
video_embeddings = np.load('/workspace/data/crops_and_embeddings/cameratrap_02_2024/embeddings/embeddings.npy')
# bristol 

embeddings = np.concatenate([bwindi_embeddings, loango_embeddings, kwitonda_embeddings, kwitonda_database_embeddings, video_embeddings])

labels = ["images"] * (len(embeddings) - len(video_embeddings)) + ["video"] * len(video_embeddings)

labels = np.array(labels)
```

```python

mapper = umap.UMAP().fit(embeddings)

umap.plot.points(mapper, labels=labels, width=900, height=700
```
