```python
import os
import matplotlib.pyplot as plt

d = {}
basepath = "../embedding-approach/data/cropped/"

for individual in os.listdir(basepath):
    path = os.path.join(basepath, individual)
    if os.path.isdir(path):
        d[individual] = len(os.listdir(os.path.join(path)))

plt.bar(range(len(d)), list(d.values()), align='center')
plt.xticks(range(len(d)), list(d.keys()))

print("Total individuals", len(d))
print("Individuals with more than 2 samples", sum([1 for val in d.values() if val > 2]))
plt.show()
```
